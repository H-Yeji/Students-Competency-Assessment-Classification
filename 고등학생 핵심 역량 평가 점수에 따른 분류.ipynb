# Training 데이터 EDA

```python
#수업 기록 데이터 불러오기
df1 = pd.read_csv('/content/5.수업기록 데이터.csv', encoding='cp949')
df1.head()
```

```python
#필요없는 컬럼 제거
df1.drop(["record_id", "mission_id", "student_performance", "student_assessment", "date"], axis=1, inplace=True)
```

```python
df1.shape
```

```python
df1.head()
```

```python
#df1_tmp로 복사
df1_tmp = df1
```

```python
df1_tmp.shape
```

```python
df1_tmp.head()
```

```python
#학생정보 데이터 (고등) 불러오기
df2 = pd.read_csv('/content/1.학생정보데이터.csv', encoding='cp949')
df2.head()
```

```python
df2.shape
```

```python
# 학생 2000명 랜덤 추출
df2_random = df2.sample(n=2000, replace=False)
df2_random.head()
```

```python
#index 순서 0부터 시작하게 설정
df2_random.reset_index(inplace=True)
```

```python
df2_random.head()
```

```python
#새로운 데이터프레임 생성
df = pd.DataFrame(columns=["student_id", "student_performance", "competency_label_1", "competency_label_2", "competency_label_3", "competency_label_4"])
df
```

```python
#df2_random에 있는 student_id가 같은 학생의 역량 정보를 df1_tmp에서 찾아서 df에 붙이기
for i in range(len(df2_random["student_id"])):
  num = df2_random.loc[i, "student_id"] #df2_random의 첫번째 학생부터 마지막 학생까지의 student_id를 저장
  tmp = df1_tmp[df1_tmp["student_id"] == num] #저장한 student_id와 같은 학생의 역량정보 데이터를 찾아서 저장
  df = pd.concat([df, tmp])
```

```python
#필요없는 컬럼 제거
df.drop(["student_performance"], axis=1, inplace=True)
```

```python
df
```

```python
#타입 확인
df.info()
```

```python
#object -> int 변환
df["competency_label_1"] = df["competency_label_1"].astype(int)
df["competency_label_2"] = df["competency_label_2"].astype(int)
df["competency_label_3"] = df["competency_label_3"].astype(int)
df["competency_label_4"] = df["competency_label_4"].astype(int)
```

```python
#df info 다시 확인 -> int로 변환된거 확인
df.info()
```

```python
df.head(15)
```

```python
#중복되는 student_id를 각 역량 점수의 평균으로 매핑
df_mean = df.groupby('student_id').mean()
df_mean.head(15)
```

```python
#크기 확인
df_mean.shape
```

```python
# 소수점 버리고 정수로 변환 -> 점수는 소수점으로 반올림하지 않기때문
df_mean = df_mean.astype(int)
df_mean
```

```python
#타입 확인
df_mean.info()
```

```python
#인덱스 0부터 시작
df_mean.reset_index(inplace=True)
```

```python
df_mean.info()
```

```python
X_train = df_mean
X_train #X_train 데이터 생성
```

# 예측 데이터(컬럼) 만들기

```python
# y 예측 데이터 만들기
X_train
```

```python
X_train['pred'] = (X_train['competency_label_1'] + X_train['competency_label_2'] +  X_train['competency_label_3'] + X_train['competency_label_4']) / 4
```

```python
X_train.info()
```

```python
#pred column: float->int 변환 (점수는 정수만 존재)
X_train['pred'] = X_train['pred'].astype(int)
```

```python
X_train.head()
```

```python
#pred 매핑 >> 1,2점: '미흡' / 3점: '보통' / 4점: '우수'
mapping = {1:"미흡", 2:"미흡", 3:"보통", 4:"우수", 5:"우수"}
X_train['pred'] = X_train['pred'].map(mapping)
```

```python
X_train
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X_train['pred'] = le.fit_transform(X_train['pred'])
```

```python
X_train['pred'].value_counts()
```

```python
X_train
```

# train data와 test data 분리

```python
# 데이터 분할
X = X_train.drop('pred', axis=1)
y = X_train[['pred']]
```

```python
X.shape, y.shape
```

```python
#훈련세트와 테스트세트 분리
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1868)
```

```python
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
X_test.shape
```

# 모델 생성 후 훈련, 예측

```python
#RandomForestClassifier를 이용한 다중분류
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score

rf_model = RandomForestClassifier(n_estimators=300, max_depth=10, n_jobs=-1, oob_score=True)
rf_model.fit(x_train.values, y_train.values)
rf_pred = rf_model.predict(x_valid) #x_valid로 예측
```

```python
rf_pred
```

```python
#정확도 확인
accuracy_score(rf_pred, y_valid)
```

```python
#X_test로 예측
pred = rf_model.predict(X_test)
pred
```

```python
#인코딩한 값을 다시 변환해서 미흡, 보통, 우수로 바꾸기
pred = le.inverse_transform(pred)
pred
```
